```python
# Importing the libraries

import pandas as pd
import zipfile
from sklearn.svm import SVC
from sklearn import metrics
import pickle
%matplotlib inline
```

```python
# Downloading Dataset from Kaggle

!kaggle competitions download digit-recognizer
```

```python
# Extracting the files from .zip

with zipfile.ZipFile('digit-recognizer.zip','r') as myzip:
    myzip.extractall()
```

```python
df = pd.read_csv('train.csv')
df.sample(10)
```

```python
df.shape
```

```python
df.info()
```

```python
# Dividing into training and test sets

X = df.drop(['label'], axis = 1)
Y = df['label']
```

```python
#Splitting the dataset into test and training set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Declaring Classifier

clf = SVC(kernel = 'rbf', random_state = 0)
clf.fit(X_train,Y_train)
```

```python
# Saving model as .pkl file

pickle.dump(clf,open('model.pkl','wb'))
```

```python
model = pickle.load(open('model.pkl','rb'))
```

```python
# Predicting the output

pred = model.predict(X_test)
```

```python
# Accuracy of the model

print('Model Accuracy is:', metrics.accuracy_score(Y_test,pred))
```
