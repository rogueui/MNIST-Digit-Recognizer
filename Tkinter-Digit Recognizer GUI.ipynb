```python
from tkinter import *
from tkinter import messagebox
import pyscreenshot as ImageGrab
import cv2
from PIL import Image
import pickle
import time
import sys

canvas_height = 180
canvas_width = 230

folder_loc = "C:/Users/Sarthak/Desktop/Mnist Dataset"

def paint(event):
    color = 'white'
    x1,y1 = (event.x - 8),(event.y - 8)
    x2,y2 = (event.x + 8),(event.y + 8)
    canvas.create_rectangle(x1,y1,x2,y2,fill = color,outline = color)
    
def delete_lines(event):
    canvas.delete('all')
    
def predict():
    
    x = root.winfo_rootx() + canvas.winfo_x()
    y = root.winfo_rooty() + canvas.winfo_y()
    x1 = x + canvas.winfo_width()
    y1 = y + canvas.winfo_height()
    #img = ImageGrab.grab(bbox=(x,y,x1,y1))
    img = ImageGrab.grab(bbox=(0, 50, 290, 260)) # X,Y,X1,Y1
    img.save(folder_loc + "/img/test_orig.png")
    
    model = pickle.load(open('model.pkl','rb'))
    
    # Converting the image into grayscale
    image = cv2.imread(folder_loc + "/img/test_orig.png")
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Converting the image into black & white
    (thresh, bnw_img) = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    cv2.imwrite(folder_loc + "/img/test_orig.png",bnw_img);
    
    # Converting the image size into 28x28 pixels
    image = Image.open(folder_loc + "/img/test_orig.png")
    new_image = image.resize((28,28))
    new_image.save(folder_loc + "/img/test_orig1.png")
    
    # Converting the values into predictable format
    im = Image.open(folder_loc + "/img/test_orig1.png")
    width, height = im.size
    pixel_values = list(im.getdata())
    
    Prediction = model.predict([pixel_values])
    Prediction = str(Prediction[0])
    
    messagebox.showinfo('Prediction','The Predicted Digit is: '+Prediction)

root = Tk()
root.title('Guess Number')

canvas = Canvas(root,height = canvas_height,width = canvas_width, bg = 'black')
canvas.pack(expand = YES,fill = BOTH)
canvas.bind('<B1-Motion>', paint)
canvas.bind('<Button-3>', delete_lines)

canvas.update()

button = Button(root,text = 'Predict',width = 40,command = predict)
button.pack(side = BOTTOM)

msg = Label(root,text = 'Press and Drag to Draw')
msg.pack(side = BOTTOM)

root.mainloop()
```
